#**Habilitando Google Colab**

```
from pycaret.utils import enable_colab
enable_colab()
```

#**Configurando PyCaret**

```
from pycaret.clustering import *
```

```
configuracion = setup(data, session_id = 123)
```

# **CreaciÃ³n del modelo**

```
models()
```

```
km = create_model('kmeans', num_clusters=5)
print(km)
```

#**Evaluando el modelo**

```
evaluate_model(km)
```

#**Asignando el modelo**

```
km_resultados = assign_model(km)
km_resultados.head()
```

#**VisualizaciÃ³n del modelo**
