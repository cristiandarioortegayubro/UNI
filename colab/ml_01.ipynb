## **Conjuntos de datos**

```
from pycaret.datasets import get_data
index = get_data('index')
```

##**Carga de un conjunto de datos determinado**

***Creación del DataFrame para un conjunto de datos determinado, como por ejemplo el conjunto de datos "insurance"***

```
data = get_data('automobile')
```

#**Configurando PyCaret**

```
from pycaret.regression import *
```

```
configuracion = setup(data, 
                      target="price", 
                      session_id = 123)
```

# **Creación del modelo**

```
compare_models()
```

```
modelo = create_model('rf')
print(modelo)
```

#**Optimizando el modelo**

```
modelo_optimo = tune_model(modelo)
```

#**Evaluando el modelo**

```
evaluate_model(modelo_optimo)
```

#**Predicción**

```
predict_model(modelo_optimo)
```

```
predict_model(modelo_optimo);
```

#**Finalización del modelo**

```
modelo_final = finalize_model(modelo_optimo)
```

```
print(modelo_final)
```

```
predict_model(modelo_final)
```

#**Guardando el modelo**

```
save_model(modelo_final,'Modelo final RF 032022')
```

#**Leyendo el modelo**

```
modelo_guardado = load_model('Modelo final RF 032022')
```

```
nueva_prediccion = predict_model(modelo_guardado, data=data)
```
