# **Conjunto de datos**

```python
from pycaret.datasets import get_data
```

```python
datos = get_data("automobile")
```

```python
datos.shape
```

## Subdivision del conjunto de datos

```python
df = datos.sample(frac=0.8, random_state=2022)
```

```python
dfnovistos = datos.drop(df.index)
```

```python
dfnovistos.drop(columns="price", inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
dfnovistos.reset_index(drop=True, inplace=True)
```

```python
print(f"Datos para el modelo {df.shape}")
print(f"Datos nuevos para predecir {dfnovistos.shape}")
```

# **Configuracion**

```python
from pycaret.regression import *
```

```python
configuracion = setup(df, 
                      target="price",
                      normalize= True,
                      remove_outliers= True, 
                      session_id=123)
```

# **Modelo**

```python
compare_models()
```

```python
modelo = create_model("lr")
```

# **Evaluacion**

```python
evaluate_model(modelo)
```

# **Prediccion**

```python
predict_model(modelo)
```

```python
predict_model(modelo);
```

# **Finalizacion**

```python
modelo_final = finalize_model(modelo)
```

```python
print(modelo_final)
```

```python
predict_model(modelo_final)
```

# **Guardar el modelo**

```python
save_model(modelo_final,"modelobase")
```
